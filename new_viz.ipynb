# Проблеми початкової візуалізації
Поточна візуалізація має незручний розподіл на роки та регіони, складно порівняти структуру ВДВ у різних регіонах. Також, складно читаються надписи на осі X. Проте, візуалізація структури ВДВ за допомогою bar chart мені здається досить вдалою та зрозумілою.

![alt text](original_viz.png)

```python
df = pd.read_excel("1.4_02.xlsx", index_col=None, usecols="A:F")
df[['Регіон', 'Рік']] = df['Регіон, рік'].str.split(",",expand=True)
df
```

```python
df_household = df[["Регіон", "Рік","сільське господарство" ]]
df_household.loc[:, 'Галузь'] = "сільське господарство"
df_household.rename({'сільське господарство': 'Разом'}, axis=1,inplace=True)

df_industry =  df[["Регіон", "Рік","промисловість" ]]
df_industry.loc[:, 'Галузь'] = "промисловість"
df_industry.rename({'промисловість': 'Разом'}, axis=1,inplace=True)

df_building= df[["Регіон", "Рік","будівництво" ]]
df_building.loc[:, 'Галузь'] = "будівництво"
df_building.rename({'будівництво': 'Разом'}, axis=1,inplace=True)

df_sevices = df[["Регіон", "Рік","послуги" ]]
df_sevices.loc[:, 'Галузь'] = "послуги"
df_sevices.rename({'послуги': 'Разом'}, axis=1,inplace=True)

df_governance = df[["Регіон", "Рік","державне управління" ]]
df_governance.loc[:, 'Галузь'] = "державне управління"
df_governance.rename({'державне управління': 'Разом'}, axis=1,inplace=True)

final_df = pd.concat([df_household, df_industry, df_building, df_sevices, df_governance])


```

```python
final_df
```

```python
final_df =final_df.sort_values(by=['Рік'], ascending=True)
zaporizhia = final_df.loc[final_df['Регіон'] == 'Запорізька']
zaporizhia.head(5)
```

```python
final_2010 = final_df.head(125)
final_2010 = final_2010.append(zaporizhia.head(5))
```

```python
final_2017 = final_df.tail(135)
final_2017 = final_2017[final_2017.Регіон != 'Запорізька']
final_2017 = final_2017.append(zaporizhia.tail(5))
```

```python
left = alt.Chart(final_2010).mark_bar().encode(
    y = alt.Y('Регіон:N', axis=None),
    x = alt.X('Разом:Q', aggregate = 'sum', sort='descending', title='Структура ВДВ, %', scale=alt.Scale(domain=[0, 100])),
    color = alt.Color('Галузь:N'),
    tooltip = alt.Tooltip('Разом:Q', aggregate = 'sum')
).properties(width = 750, height = 500, title = '2010')

left
```

```python
right = alt.Chart(final_2017).mark_bar().encode(
    y = alt.Y('Регіон:N', axis=None),
    x = alt.X('Разом:Q', aggregate = 'sum', title='Структура ВДВ, %', scale=alt.Scale(domain=[0, 100])),
    color = alt.Color('Галузь:N'),
    tooltip = alt.Tooltip('Разом:Q', aggregate = 'sum')
).properties(width = 750, height = 500, title = '2017')
right
```

```python
axis = alt.Chart(final_2017).encode(
    y=alt.Y('Регіон:N', axis=None),
    text=alt.Text('Регіон:N'),
).mark_text().properties(width=20,height = 500 )
axis
```

```python
alt.concat(left, axis, right, spacing=5).properties(background = '#F9F9F9', padding = 25, title = "Зміна структури ВДВ у розрізі регіонів за 2010 та 2017 роки").configure_title(fontSize=20)
```
